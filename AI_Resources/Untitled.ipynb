```python
import tensorflow as tf
import pandas as pd
import numpy as np

# Define the first AI model
def create_model_2(input_shape):
    model = tf.keras.Sequential([
        Dense(64, activation='relu', input_shape=input_shape),
        Dense(32, activation='relu'),
        Dense(4, activation='softmax'),  # Use softmax for multi-class classification
    ])
    return model

# Load your data
data_2 = pd.read_csv("data_2.csv")

# Assuming "label" is the column you want to predict
labels_2 = data_2["Label"]

# Convert non-numeric labels to numerical format (one-hot encoding)
unique_labels = labels_2.unique()
label_mapping = {label: i for i, label in enumerate(unique_labels)}
numeric_labels = labels_2.map(label_mapping)

# Create the first AI model
model_2 = create_model_2((5,))

# Extract input data
input_data_2 = data_2.drop("Label", axis=1)

# Compile and train the model
model_2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_2.fit(input_data_2.values, numeric_labels, epochs=40)

# Now, you can use the trained model to make predictions
# Assuming you have a new dataset for prediction
new_data = np.array([[0.79042,0.83232,0.408368,0.604073,0.370097]])

# Make predictions using the trained model
predictions = model_2.predict(new_data)

# Get the predicted class
predicted_class = np.argmax(predictions)

# Get the predicted label
predicted_label = unique_labels[predicted_class]

# Display the predictions
print("Predictions:", predictions)   
print("Predicted Class:", predicted_class)
print("Predicted Label:", predicted_label)
```
